```python
import numpy as np
import pandas as pd
import scipy.io
np.random.seed(64)
XYZ_Coord = scipy.io.loadmat('data\XYZcoord1.mat')
XYZ_C = np.array(XYZ_Coord ['XYZ1'])
XYZ_C
```

```python
XYZ_Edges = pd.read_csv('data\Edges.csv')
XYZ_Edges
```

```python
XYZ_E = np.array(XYZ_Edges.iloc[:, 0:3].values)
XYZ_E
```

```python
Euclidean_distance = []

# generate Euclidean Distance from data coordinate
for i in range(np.size(XYZ_C,0)):
    for j in range(i+1, len(XYZ_C)):
        temp = 0
        for k in range(np.size(XYZ_C,1)):
            temp = temp + (XYZ_C[i][k] - XYZ_C[j][k])**2
        Euclidean_distance.append([i+1, j+1 ,temp**-0.5])
np.array(Euclidean_distance)
Euclidean_distance
```

```python
#Use the following command to install required package.
#conda install python-louvain networkx
#https://python-louvain.readthedocs.io/en/latest/index.html
#https://networkx.github.io/documentation/stable/index.html
import community
import networkx as nx
import matplotlib.pyplot as plt

#Use pre-defined linkage (Edges.csv) to constructure whole network
G_geodesic = nx.Graph()
for i in range(0, len(XYZ_E)):    
    e = ( str(int(XYZ_E[i,0])), str(int(XYZ_E[i,1])), XYZ_E[i,2] )
    G_geodesic.add_weighted_edges_from([(e)])

G_euclidean = nx.Graph()
for j in range(0, len(Euclidean_distance)):    
    e = ( str(int(Euclidean_distance[j][0])), str(int(Euclidean_distance[j][1])), Euclidean_distance[j][2] )
    G_euclidean.add_weighted_edges_from([(e)])
```

```python
partition = community.best_partition(G_geodesic)
size = float(len(set(partition.values())))
mod = community.modularity(partition,G_geodesic)
for i in range(5):
    partition_2 = community.best_partition(G_geodesic)
    size_2 = float(len(set(partition.values())))
    mod_2 = community.modularity(partition,G_geodesic)
    if mod_2 > mod:
        partition = partition_2
        size = size_2
        mod = mod_2
    print(i)

partition2 = community.best_partition(G_euclidean)
size2 = float(len(set(partition2.values())))
mod2 = community.modularity(partition2,G_euclidean)
for j in range(5):
    partition_2 = community.best_partition(G_euclidean)
    size_2 = float(len(set(partition.values())))
    mod_2 = community.modularity(partition,G_euclidean)
    if mod_2 > mod2:
        partition2 = partition_2
        size2 = size_2
        mod2 = mod_2
    print(j)
```

```python
print("community_geodesic:", size)
print("modularity_geodesic:", mod)
print("community_euclidean:", size2)
print("modularity_euclidean:", mod2)
```

```python
#assign node color based on geodesic community in network
display('Geodesic community')
for com in set(partition.values()) :
    print("Community", com)
    members = list_nodes = [nodes for nodes in partition.keys() if partition[nodes] == com]
    print(members)
values = [partition.get(node) for node in G_geodesic.nodes()]
nx.draw_spring(G_geodesic, cmap = plt.get_cmap('jet'), node_color = values, node_size = 30, with_labels = False)
plt.show()

#assign node color based on Euclidean community in network
display('Euclidean community')
for com2 in set(partition2.values()) :
    print("Community", com2)
    members2 = list_nodes2 = [nodes2 for nodes2 in partition2.keys() if partition2[nodes2] == com2]
    print(members2)
values2 = [partition2.get(node2) for node2 in G_euclidean.nodes()]
nx.draw_spring(G_euclidean, cmap = plt.get_cmap('jet'), node_color = values2, node_size = 30, with_labels = False)
plt.show()
```

```python
#assign data point color based on geodesic community in realspace
label = np.zeros((len(XYZ_C),1))
for j in set(partition.values()) :
    for i in range(len(XYZ_C)) :
        if partition[str(i+1)] == j :            
            label[i] =  j
labelRE = np.reshape(label, len(XYZ_C))      

#assign data point color based on Euclidean community in realspace
label2 = np.zeros((len(XYZ_C),1))
for j in set(partition2.values()) :
    for i in range(len(XYZ_C)) :
        if partition2[str(i+1)] == j :            
            label2[i] =  j
labelRE2 = np.reshape(label2, len(XYZ_C))
```
